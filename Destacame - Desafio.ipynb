## Explicación de las diferencias entre los promedios de los calculos en las tablas 



Los montos promedios adeudados y a pagar se estan calculando a diferentes niveles de granularidad. Es decir, en el dataframe de "users" se calcula a nivel de usuario y "receivers" a nivel de institución. Ademas cabe resaltar que 
la media (promedio) no es una medida robusta debido a su sensibilidad a los valores atípicos o extremos en un conjunto de datos. Un valor atípico es una observacion que es numericamente distante al resto de los datos y eso puede generar que los resultados se vean afectados.
